```python
# lenskart

from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np

Product_name =[]
Product_price =[]
Product_offer =[]
Product_rating = []

url = "https://www.lenskart.com/eyeglasses/collections/all-computer-glasses.html#frame_color_id=13323"    
r =  requests.get(url)
print(r)

soup = BeautifulSoup(r.text,'lxml')
#print(soup)

names = soup.find_all("div",class_ = "product-name")
for i in names:
    name = i.text
    Product_name.append(name)
print(Product_name)
print(len(Product_name))

names = soup.find_all("span",class_ = "text-color-dark-blue fw700")
for i in names:
    name = i.text
    Product_price.append(name)
print(Product_price)
print(len(Product_price))

names = soup.find_all("span",class_ = "product-offer-text")
for i in names:
    name = i.text
    Product_offer.append(name)
print(Product_offer)
print(len(Product_offer))

names = soup.find_all("span",class_ = "rating-value")
for i in names:
    name = i.text
    Product_rating.append(name)
print(Product_rating)
print(len(Product_rating))

df = pd.DataFrame({"Product Name":Product_name, "price":Product_price, "offer":Product_offer, "rating":Product_rating})
print(df)


df.to_csv("lenskart.csv")
df
```
